# Instructions for Downloading the Dataset
This project uses the **SVHN_single_grey1.h5** file from the Google Street View House Numbers Dataset, which is available on Kaggle. To download the dataset, please follow these steps:

1. Go to the [Google Street View House Numbers Dataset on Kaggle](https://www.kaggle.com/datasets/iarunava/cell-images-for-detecting-malaria).
2. Download the **SVHN_single_grey1.h5** file to your local machine.
3. Save the downloaded file to a directory you can easily access.
4. Update your code or notebook with the path to the directory where you saved the file.
