# Instructions for Downloading the Dataset
This project uses the **Malaria Cell Images Dataset**, available on Kaggle. To download the dataset, please follow these steps:
1.	Navigate to the [Malaria Cell Images Dataset on Kaggle](https://www.kaggle.com/datasets/iarunava/cell-images-for-detecting-malaria).
2.	 Download the contents of the `cell_images` directory, which consists of two subdirectories.
3.	 Extract the downloaded files and save them to a directory you can easily access.
4.	 Make sure to update your code or notebook with the path to the directory where you saved the files.
